```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import os
import numpy as np
from pathlib import Path
from collections import Counter
```

```python
# Read the CSV file into a Pandas DataFrame
file_path = Path('COVID-19_Case_Surveillance_Public_Use_Data_with_GeographyCLEAN.csv')
df = pd.read_csv(file_path)
```

```python
df.head()
```

```python
df = df.replace("No",0)
```

```python
df = df.replace("Yes",1)
```

```python
df.head()
```

```python
pd.get_dummies(df, columns=["sex"])
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df2 = df.copy()
df2['age_group'] = le.fit_transform(df2['age_group'])
df2['sex'] = le.fit_transform(df2['sex'])

df2=df2.drop(columns=["case_month", "res_state", "res_county", "race", "ethnicity", "case_positive_specimen_interval", "case_onset_interval", "process", "exposure_yn", "current_status","symptom_status"])
```

```python
df2
```

```python
#remove any last strings
df2 = df2[~df.isin(['Missing', 'Unknown']).any(axis=1)]
df2 = df2.dropna()
```

```python
df2
```

```python
# Define the features set.
y = df2["death_yn"]
#X = df2["underlying_conditions_yn"]
X = df2.drop(columns="death_yn")
X.head()
```

```python
# Define the target set.
y = df2["death_yn"].values
y[:5]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Determine the shape of our training and testing sets. 75% 25% split
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
# Creating a StandardScaler instance.
scaler = StandardScaler()
# Fitting the Standard Scaler with the training data.
X_scaler = scaler.fit(X_train)

# Scaling the data.
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
from sklearn import tree

# Creating the decision tree classifier instance.
model = tree.DecisionTreeClassifier()
# Fitting the model.
model = model.fit(X_train_scaled, y_train)
```

```python
# Making predictions using the testing data.
predictions = model.predict(X_test_scaled)
```

```python
predictions
```

```python
# Calculating the confusion matrix
cm = confusion_matrix(y_test, predictions)

# Create a DataFrame from the confusion matrix.
cm_df = pd.DataFrame(
    cm, index=["Actual 0", "Actual 1"], columns=["Predicted 0", "Predicted 1"])

cm_df
```

```python
# Calculating the accuracy score.
acc_score = accuracy_score(y_test, predictions)
```

```python
# Displaying results
print("Confusion Matrix")
display(cm_df)
print(f"Accuracy Score : {acc_score}")
print("Classification Report")
print(classification_report(y_test, predictions))
```
