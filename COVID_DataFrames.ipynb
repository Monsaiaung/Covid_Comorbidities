```python
# Load in CSV files
file_path = 'Resources/Covid-19/Data/Provisional_COVID-19_Deaths_by_Sex_and_Age.csv'
Covid_sex = pd.read_csv(file_path)
Covid_sex
```

```python
# Dropped Data as Of column (only states date data was downloaded)
# Dropped footnote, not much useful information
Covid_sex.drop(columns=['Footnote', 'Data As Of'], inplace=True)
```

```python
Covid_sex.head(5)
```

```python
# Selecting subset By total
Covid_sex_by_total = Covid_sex.loc[Covid_sex['Group'] == 'By Total']
Covid_sex_by_total
```

```python
# Dropping year and month
Covid_sex_by_total.drop(columns=['Year','Month'], inplace=True)
Covid_sex_by_total
```

```python
# Combining start and end date
Covid_sex_by_total['Date'] = Covid_sex_by_total['Start Date'] + '-' + Covid_sex_by_total['End Date']
Covid_sex_by_total
```

```python
# Set Date as index, dropped start and end date
Covid_sex_by_total.drop(columns=['Start Date','End Date'], inplace=True)
Covid_sex_by_total.set_index('Date', drop=True, inplace=True)
Covid_sex_by_total
```

```python
# Drop Group column
Covid_sex_by_total.drop(columns=['Group'], inplace=True)
Covid_sex_by_total.head(5)
```

```python
# We have to decide how to deal with NaN's without distorting the data. Should we replace some with zeros, or remove rows entirely?
# Most NaNs come from the really young groups which suggests that a zero might be appropriate
Covid_sex_by_total.loc[Covid_sex_by_total['State'] == 'New York']
```

```python
Covid_sex.tail(5)
```

```python
# Creating DataFrame by month
Covid_sex_by_month = Covid_sex.loc[Covid_sex['Group'] == 'By Month']
Covid_sex_by_month
```

```python
# Dropping redundant columns
Covid_sex_by_month.drop(columns=['End Date','Group'], inplace=True)
Covid_sex_by_month.drop(columns=['Year', 'Month'], inplace=True)
Covid_sex_by_month.head(5)
```

```python
# Converting Object to Date
Covid_sex_by_month['Start Date'] = pd.to_datetime(Covid_sex_by_month['Start Date'])
```

```python
# Converting date to month year format
Covid_sex_by_month['Start Date'] = Covid_sex_by_month['Start Date'].dt.strftime('%d-%Y')
Covid_sex_by_month.tail()
```

```python
df = Covid_sex_by_month.loc[Covid_sex_by_month['Age Group'] == 'All Ages']
df = df.loc[df['Sex'] == 'All Sexes']

# Covid_sex_by_month.groupby(['Year', 'Month']).sum()['Total Deaths']
```

```python
deaths_per_month_df = df.loc[df['State'] == 'United States'].groupby(['Year', 'Month']).sum()[['COVID-19 Deaths', 'Pneumonia Deaths', 'Total Deaths']]
deaths_per_month_df
```

```python
# Export csv file
filepath = 'Resources/Covid-19/Cleaned_Data/Total_Deaths_per_Month.csv'
deaths_per_month_df.to_csv(filepath)
```
