```python
import pandas as pd
```

```python
file_path = 'Resources/Covid-19/Data/Rates_of_COVID-19_Cases_or_Deaths_by_Age_Group_and_Vaccination_Status.csv'
Vaccination_df = pd.read_csv(file_path)
Vaccination_df
```

```python
Vaccination_df['Month'] = pd.to_datetime(Vaccination_df['month'])
```

```python
Vaccination_df['Month'] = Vaccination_df['Month'].dt.strftime('%d-%Y')
```

```python
Vaccination_df.columns.tolist()
```

```python
Vaccination_df.dtypes
```

```python
Jab_df = Vaccination_df[['Month','outcome',
                'MMWR week',
                 'Age group',
                 'Vaccine product',
                 'Vaccinated with outcome',
                 'Fully vaccinated population',
                 'Unvaccinated with outcome',
                 'Unvaccinated population']]
```

```python
Jab_df
```

```python
Jab_df = Jab_df.loc[Jab_df['Vaccine product'] == 'all_types']
```

```python
Jab_df.groupby(['outcome','Age group']).sum([['Vaccinated with outcome', 'Unvaccinated with outcome']])
```

```python
Vax_vs_unvax = Jab_df.groupby(['Age group','outcome']).sum()[['Vaccinated with outcome', 'Unvaccinated with outcome']]
```

```python
file_path = 'Resources/Covid-19/Cleaned_Data/Vax_vs_unvax.csv'
Vax_vs_unvax.to_csv(file_path)
```
