```python
# Imports
import pandas as pd
import numpy as np
import csv
import datetime as dt
```

```python
# Load in CSV files
file_path = 'Resources/Conditions_Contributing_to_COVID-19_Deaths__by_State_and_Age__Provisional_2020-2022.csv'
Covid_contributing_conditions = pd.read_csv(file_path)
Covid_contributing_conditions
```

```python
Covid_contributing_conditions.columns.tolist()
```

```python
Covid_contributing_conditions.drop(columns=['Data As Of', 'Flag'], inplace=True)
```

```python
Covid_contributing_conditions.head()
```

```python
# Selecting subset By All Age in Age Group
Covid_contributing_conditions_Allage = Covid_contributing_conditions.loc[Covid_contributing_conditions['Age Group'] == 'All Ages']
Covid_contributing_conditions_Allage
```

```python
# Selecting second subset By Total in Group 
Covid_contributing_conditions_ByTotal = Covid_contributing_conditions_Allage.loc[Covid_contributing_conditions_Allage['Group'] == 'By Total']
Covid_contributing_conditions_ByTotal
```

```python
# Dropping year and month
Covid_contributing_conditions_ByTotal.drop(columns=['Year','Month'], inplace=True)
Covid_contributing_conditions_ByTotal
```

```python
# Combining start and end date
Covid_contributing_conditions_ByTotal['Date'] = Covid_contributing_conditions_ByTotal['Start Date'] + '-' + Covid_contributing_conditions_ByTotal['End Date']
Covid_contributing_conditions_ByTotal
```

```python
# Set Date as index, dropped start and end date
Covid_contributing_conditions_ByTotal.drop(columns=['Start Date','End Date'], inplace=True)
Covid_contributing_conditions_ByTotal.set_index('Date', drop=True, inplace=True)
Covid_contributing_conditions_ByTotal
```

```python
# Create a DataFrame object
Covid_contributing_conditions_df = pd.DataFrame(Covid_contributing_conditions_ByTotal, columns= ['Group', 'State','Condition Group', 'Condition', 'ICD10_codes', 'Age Group', 'COVID-19 Deaths', 'Number of Mentions'])
```

```python
print(Covid_contributing_conditions_df)
```

```python
# Export csv file
filepath = 'Resources/Covid_contributing_conditions.csv'
Covid_contributing_conditions_df.to_csv(filepath)
```
